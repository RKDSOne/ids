## Array 2 Series
Performance from `Array` to `Series`.

```python
def array2series(n):
    ori=np.random.randn(n)
    tpre=datetime.now()
    tst=pd.Series(ori)
    return (datetime.now()-tpre).total_seconds()


x=[20000000*i for i in xrange(1,16)]
y=[]
for i in x:
    y.append(array2series(i))
plt.gcf().set_size_inches(8,5)
plt.plot(x, y)
plt.ylabel(r'Time Consumed $t(s)$')
plt.xlabel(r'Array Size $n$')
```

## Series Value Counts
Performance of `pandas.Series.value_counts()`

```python
def tictoc_valuecounts(n):
    ori=[int(i) for i in np.random.rand(n)]
    sr=pd.Series(ori)
    tpre=datetime.now()
    res=sr.value_counts()
    return (datetime.now()-tpre).total_seconds()


x=[5000000*i for i in xrange(1,16)]
y=[]
for i in x:
    y.append(tictoc_valuecounts(i))
plt.gcf().set_size_inches(8,5)
plt.plot(x, y)
plt.ylabel(r'Time Consumed $t(s)$')
plt.xlabel(r'Array Size $n$')
```

## Efficiency of `numpy.random.shuffle()`
Test the efficiency infulence from unit size of `numpy.random.shuffle`.
Test in size 1000000. with unit size in [1,3,10,30,100,300,1000]

```python
from datetime import *
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline

def foo(n):
    X=random.rand(1000000, n)
    t1=datetime.now()
    random.shuffle(X)
    return (datetime.now()-t1).total_seconds()


x=range(10,520,50)
y=[]
for i in x:
    y.append(foo(i))
plt.plot(x,y)
plt.show()
```

## Vectorized efficiency

Test performance of Batch-Sort, i.e., apply ascending sort to all rows of a matrix.
  
We would compare the time consumed between `for` loop, `list` constructor and, the **state-of-the-art** `numpy.apply_along_axis`.

```python
from datetime import *
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline


def vectorized(n):
    X=random.rand(n,100)
    t1=datetime.now()
    X = apply_along_axis(sorted, 1, X)
    return (datetime.now()-t1).total_seconds()


def list_constructor(n):
    X=random.rand(n,100)
    t1=datetime.now()
    X=[sorted(i) for i in X]
    return (datetime.now()-t1).total_seconds()


def for_loop(n):
    X=random.rand(n,100)
    t1=datetime.now()
    oriX=X
    X=[]
    for i in oriX:
        X.append(sorted(i))
    X=array(X)
    return (datetime.now()-t1).total_seconds()


print vectorized(1000000)
print list_constructor(1000000)
print for_loop(1000000)
```
